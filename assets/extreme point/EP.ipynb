Algorithm 1: EP-INSERT

Input:
    B    // a bin (container)
    I    // an item to be packed

Output:
    placement or failure

1:  EP ← initial extreme point set of bin B
2:  for each extreme point p ∈ EP do
3:      for each allowed orientation o of item I do
4:          if I can be placed at p with orientation o
5:             without overlapping existing items then
6:              compute placement score at (p, o)
7:              store feasible placement (p, o)
8:  if feasible placements ≠ ∅ then
9:      select best placement according to heuristic score
10:     place item I into bin B at selected (p, o)
11:     update extreme point set EP
12:     return placement
13: else
14:     return failure


Algorithm 2: EP-INSERT-SD (Space Defragmentation)

Input:
    B    // a bin
    I    // an item

Output:
    placement or failure

1:  generate feasible placements using EP-INSERT
2:  for each feasible placement do
3:      evaluate space defragmentation quality
4:      // e.g., size and regularity of remaining free space
5:  if feasible placements ≠ ∅ then
6:      select placement that best improves space compactness
7:      place item I into bin B
8:      update extreme point set EP
9:      return placement
10: else
11:     return failure



Algorithm 3: BIN-SHUFFLE-IMPROVE

Input:
    S    // initial packing solution (set of bins)
Output:
    improved solution S'

1:  repeat
2:      select one or more bins from S
3:      remove all items from selected bins
4:      reorder removed items
5:      reinsert items using EP-INSERT or EP-INSERT-SD
6:      if new solution uses fewer bins or is more compact then
7:          update S
8:  until no improvement is found
9:  return S


| 缩写              | 全称                  | 含义             |
| ---------------- | --------------------- | --------------- |
| **EP**           | Extreme Point         | 角点/极点：物品可放置的候选点 |
| **Corner Point** | –                     | EP 的早期称呼，本质相同   |
| **SD**           | Space Defragmentation | 空间碎片整理          |


| 符号    | 含义                        |
| ----- | ------------------------- |
| **B** | Bin / Container（容器）       |
| **I** | Item（待放置物品）               |
| **S** | Packing Solution（装箱方案）    |
| **p** | 一个 extreme point（候选角点）    |
| **o** | item 的一个允许朝向（orientation） |
